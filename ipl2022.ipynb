```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("C:\\Users\\raman\\Downloads\\archive (1)\\IPL_Matches_2022.csv")

print(data.head)
```

```python
data.columns
```

```python
print('total rows and columns:',data.shape)
```

```python
print('total matches played in year 2022:',data.shape[0])
```

```python
wins=data['WinningTeam'].value_counts()

print(wins)

#analyzing the winning data in visualization formate

wins.plot(kind='bar', color='green')
plt.title("Matches Won by Teams - IPL 2022")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=60)
plt.show()
```

```python
top_players=data['Player_of_Match'].value_counts().head(5)

print(top_players)


#analyzing the winning data in visualization formate

top_players.plot(kind='bar',color='purple')
plt.title("most man of the match award")
plt.xlabel('players')
plt.ylabel('no of wins')
plt.xticks(rotation=60)
plt.show()
```

```python
no_match=data['Venue'].value_counts()
print(no_match)

no_match.plot(kind='pie')
plt.title('no of matches at each venue')
plt.show()
```

```python
toss_decision=data['TossDecision'].value_counts()

print(toss_decision)


toss_decision.plot(kind='pie')
plt.title('toss decisions')
plt.show()
```
